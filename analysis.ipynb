```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as pyplot 
powerFile = "power.csv"
energyFile = "energy.csv"
logFile = "log.csv"
```

```python
pwr = pd.read_csv(powerFile)
engy = pd.read_csv(energyFile)
log = pd.read_csv(logFile, skiprows = 1)
pwr = pwr.drop(columns = ['SampleNumber', 'Duration'])
engy = engy.drop(columns = ['SampleNumber', 'Duration'])
pwr = pwr.rename(index=str, columns={'PowerData':'Power(mW)'})
engy = engy.rename(index=str, columns={'EnergyData':'Power(mJ)'})
```

```python
ax = pwr.plot(x='CTime')
for t in log['CTime']:
    ax.axvline(t, color='k', linestyle='--')
```

```python
ax = engy.plot(x='CTime')
for t in log['CTime']:
    ax.axvline(t, color='k', linestyle='--')
```
